# Window Function and Ranking 

```python
from pyspark.sql.window import Window 
```

```python
 
## Rank Top Selling Products Per seller 


window_spec = Window.partitionBy('seller_id').orderBy(desc('price'))
```

```python
## Rank Top Selling Products Per seller 


top_sellers_products_df = full_orders_df.withColumn('rank',rank().over(window_spec)).filter(col('rank')<=5)
top_sellers_products_df.select('seller_id','price','rank').show()
```

```python
## Dense Rank for Sellers Based on Revenue 

from pyspark.sql.functions import sum, col, dense_rank
from pyspark.sql.window import Window

# Step 1: Revenue per seller per partition (e.g., per product category)
seller_revenue = (
    full_orders_df
    .groupBy("product_category_name", "seller_id")   # partition key + seller
    .agg(sum(col("price") * col("order_item_id")).alias("total_revenue"))
)

# Step 2: Window partitioned by category, ordered by revenue
windowSpec = Window.partitionBy("product_category_name").orderBy(col("total_revenue").desc())

# Step 3: Apply dense rank inside partition
ranked_sellers = (
    seller_revenue
    .withColumn("dense_rank", dense_rank().over(windowSpec))
)

ranked_sellers.show(20)
```

# Advance Aggregation and Enrichment 

```python
full_orders_df
```

```python
## Total Revenue and Average Order Value (AOV) per Customer.

customer_spending_df = full_orders_df.groupBy('customer_id')\
.agg(
    count('order_id').alias('total_orders'),
    sum('price').alias('total_spent'),
    round(avg('price'),2).alias('AOV')
)\
.orderBy(desc('total_spent'))

customer_spending_df.shows()

```

```python
## Seller Performance Metrics(Revenue,Average Review,Order Count)
```

```python
from pyspark.sql.functions import sum,col,count,avg,round
seller_performance = full_orders_df.groupBy('seller_id')\
.agg(
    sum(col('price')* col('order_item_id')).alias('Total_revenue'),
    count('order_id').alias('Order_count'),
    round(avg('review_score'),2).alias('average_review'),## Round to 2 decimals
    round(stddev('price'),2).alias('price_variability')
    
)\
.orderBy(desc('total_revenue'))

                   
          
seller_performance.show()          
    
        
```

```python
## Product Popularity Metrics 

product_metrics_df = full_orders_df.groupBy('product_id')\
.agg(
    count('order_id').alias('total_sales'), 
    sum('price').alias('total_revenue'),
    round(avg('price'),2).alias('avg_price'), 
    round(stddev('price'),2).alias('price_volatility'),\
    collect_list('seller_id').alias('seller_list'),
    collect_set('seller_id').alias('unique_sellers')
)\
.orderBy(desc('total_sales'))
```

```python
product_metrics_df.show()
```

```python
## Monthly Revenue and Order Count Trend (To find: total_orders,total_revenue,avg_order_value,min_order_value,max_order_Value)

from pyspark.sql.functions import sum, col, count, avg, min, max, round, to_date, year, month

monthly_revenue = (
    full_orders_df
    .withColumn("order_date", to_date(col("order_purchase_timestamp")))
    .withColumn("year", year("order_date"))
    .withColumn("month", month("order_date"))
    .groupBy("year", "month")   
    .agg(
        count("order_id").alias("total_orders"),
        sum(col("price") * col("order_item_id")).alias("total_revenue"),  # use 'quantity' if available
        round(avg(col("price") * col("order_item_id")), 2).alias("avg_order_value"),
        min(col("price") * col("order_item_id")).alias("min_order_value"),
        max(col("price") * col("order_item_id")).alias("max_order_value")
    )
    .orderBy("year", "month")
)

monthly_revenue.show(20)






```

```python
 
    
##Explanation:(Above)

##to_date(order_purchase_timestamp) → converts timestamp to date.

##year, month → extract year and month for grouping.

##count(order_id) → number of orders in that month.

##sum(price*quantity) → total revenue.

##avg(price*quantity) → average order value (AOV).

##min, max → smallest and largest single order values.

##.orderBy(year, month) → chronological trend.
```

```python
## Customer Retention Analysis(First and Last Order)



customer_retention_df = full_orders_df.groupBy('customer_id')\
.agg(
    min(col('order_purchase_timestamp')).alias('first_order_Date'),
    max(col('order_purchase_timestamp')).alias('last_order_Date'),
    count('order_id').alias('total_orders'),
    round(avg('price'),2).alias('aov')
    
)\
.orderBy(desc('total_orders'))


```
